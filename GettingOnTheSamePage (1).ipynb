```python
from random import seed
from random import random
import matplotlib.pyplot as plt

seed(1)
# To generate the random data
numList = []
for i in range (0,1000):
    x = random()
    numList.append(x)

#Plotting the data set
y = plt.hist(numList, bins=25)
y = plt.xlabel('value')
y = plt.ylabel('occurance')

```

```python
from random import seed
import random
import matplotlib.pyplot as plt

seed(1)
# To generate the gaussian distribution random data
numList = []
for i in range (0,3000):
    x = random.gauss(0,2)
    numList.append(x)

#Plotting the data set
y = plt.hist(numList, bins=40)
y = plt.xlabel('value')
y = plt.ylabel('occurance')

```
